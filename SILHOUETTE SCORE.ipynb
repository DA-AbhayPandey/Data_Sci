```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Iris raw.csv")
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range (2,21):
    km=KMeans(n_clusters=i,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xlabel("No. of cluster")
plt.xticks([i for i in range(2,21)])
plt.ylabel("WCSS")
plt.grid(axis="x")
plt.show()
```

```python
kmn=KMeans(n_clusters=3)
dataset["predict"]=kmn.fit_predict(dataset)
```

```python
kmn.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(dataset,labels=kmn.labels_)
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset,hue="predict")
plt.show()
```

```python
ss = []
no_clu = [j for j in range(2,20)]
for i in range(2,20):
    km1=KMeans()
    km1.fit(dataset)
    ss.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_clu,ss)
plt.xlabel("no_clu")
plt.ylabel(ss)
plt.xticks(no_clu)
plt.grid(axis="x")
plt.show()
```

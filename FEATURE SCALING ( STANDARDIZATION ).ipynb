```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
#Now lets do scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_SS"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]))
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(11,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2,)
plt.title("After")
sns.distplot(dataset["ApplicantIncome_SS"])
plt.show()
```

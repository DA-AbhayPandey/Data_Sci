```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("Placement classification.csv")
dataset.drop(columns=["Student_ID"],inplace=True)
dataset.head()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="CGPA",y="IQ",data=dataset,hue="Placement")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=27)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100 , sv.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

# POLY

```python
svp = SVC(kernel="poly")
svp.fit(x_train,y_train)
```

```python
svp.score(x_test,y_test)*100 , svp.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svp)
plt.show()
```

# RBF

```python
svr = SVC(kernel="rbf")
svr.fit(x_train,y_train)
```

```python
svr.score(x_test,y_test)*100 , svr.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svr)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
x=dataset.drop(columns=["Salary"])
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kr = KNeighborsRegressor(n_neighbors=2)
kr.fit(x_train,y_train)
```

```python
kr.score(x_test,y_test)*100
```

```python
for i in range(1,10):
    krr = KNeighborsRegressor(n_neighbors=i)
    krr.fit(x_train,y_train)
    print(i, " - ",krr.score(x_test,y_test)*100, krr.score(x_train,y_train)*100)
```

```python
for i in range(1,50):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=i)
    krr = KNeighborsRegressor(n_neighbors=3)
    krr.fit(x_train,y_train)
    print(i, " - ",krr.score(x_test,y_test)*100, krr.score(x_train,y_train)*100)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=38)
kr38 = KNeighborsRegressor(n_neighbors=3)
kr38.fit(x_train,y_train)
```

```python
kr38.score(x_test,y_test)*100 , kr38.score(x_train,y_train)*100
```

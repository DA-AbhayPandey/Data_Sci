```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("placement.csv")
dataset.head()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa", y="package",data=dataset)
plt.show()
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict([[6.95]])
#We predicted the package of 6.95 cgpa through our train data
```

```python
# Now we check the answer with the help of formula y = x*m+c
#coef refers to m
lr.coef_ , lr.intercept_
#intercept refers to c
```

```python

6.95*0.57111357+(-0.9998951779221392)

#x = 6.95 from row 36
#m = 0.57111357 from row 38
#c = -0.9998951779221392 from row 38
```

```python
#Now check the prediction score of our train data
lr.score(x_test,y_test)*100 # *100 for get the prediction in %
```

```python
y_pre = lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.plot(dataset["cgpa"],y_pre,c="red")
plt.legend(["org_data","Pred_Line"])
plt.savefig("Simple_Linear_prediction.jpg")
plt.show()
```

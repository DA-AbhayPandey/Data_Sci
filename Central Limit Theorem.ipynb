```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = [np.random.randint(0,100) for i in range(10000)]
datatable = pd.DataFrame({"values":dataset})
```

```python
plt.figure(figsize=(4,2))
sns.kdeplot(x="values",data=datatable)
plt.show()
```

```python
#Pickup less than 10% of sample data 
sample_mean =[]
for no_sample in range(50):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(dataset))
    sample_mean.append(np.mean(sample_data))
        
sample_M = pd.DataFrame({"Sample Mean":sample_mean})
```

```python
plt.figure(figsize=(4,2))
sns.kdeplot(x="Sample Mean",data=sample_M)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,4))
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_MMS"]=pd.DataFrame(mms.transform(dataset[["CoapplicantIncome"]]))
dataset.head(5)
```

```python
plt.figure(figsize=(15,3))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_MMS"])
plt.show()
```

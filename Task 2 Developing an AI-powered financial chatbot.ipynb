```python
import pandas as pd

#importing data
df = pd.read_csv("C:/Users/ASUS/Downloads/Skills/GEN AI/BCG GenAI/SEC EDGAR.csv")
df.head()
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df.fillna(0, inplace=True)
df.head(9)
```

```python
def financial_chatbot(company_input, fiscal_year, user_query):
    company_input = company_input.capitalize()
    user_query = user_query.strip().lower()
    
    try:
        df
        
        if user_query == "what is the total revenue?":
            value = df[(df["Year"] == fiscal_year) & (df["Company"] == company_input)]["Total Revenue"].values[0]
            return f"Total Revenue for {company_input} in {fiscal_year} is ${value}"
        
        elif user_query == "what is the net income?":
            value = df[(df["Year"] == fiscal_year) & (df["Company"] == company_input)]["Net Income"].values[0]
            return f"Net Income for {company_input} in {fiscal_year} is ${value}"
        
        elif user_query == "what is the sum of total assets?":
            value = df[(df["Year"] == fiscal_year) & (df["Company"] == company_input)]["Total Assets"].values[0]
            return f"Total Assets for {company_input} in {fiscal_year} is ${value}"
        
        elif user_query == "what is the revenue growth (%) ?":
            value = df[(df["Year"] == fiscal_year) & (df["Company"] == company_input)]["Revenue Growth (%)"].values[0]
            return f"Revenue Growth for {company_input} in {fiscal_year} is {round(value, 2)}%"
        
        else:
            return "Sorry, that query is not supposed or doesn't match expected format."
        
    except IndexError:
        return "No data found for the given company and year"
    except Exception as e:
        return f"An error occurred: {str(e)}"
        
        
```

```python
#Chatbot Interaction

company = input("Enter company name from (Apple, Tesla, Microsoft): ").capitalize()
fiscal_year = int(input("Enter fiscal year between 2020 to 2024 : "))
print("Supported queries:")
print("- What is the total revenue?")
print("- What is the net income?")
print("- What is the sum of total assets?")
print("- What is the revenue growth (%)?")
query = input("Enter your query: ")

response = financial_chatbot(company, fiscal_year, query)
print("\n🤖🇦🇮 Chatbot Response:", response)
```

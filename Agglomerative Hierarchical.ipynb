```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Iris raw.csv")
dataset.head()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,10))
sch.dendrogram(sch.linkage(dataset,method='single', metric='euclidean'))
plt.savefig("dendrogram.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
dataset["Predict"]=ac.fit_predict(dataset)
dataset.head()
```

```python
sns.pairplot(dataset,hue="Predict")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("Placement classification.csv")
dataset.drop(columns="Student_ID",inplace=True)
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="CGPA",y="IQ",data=dataset,hue="Placement")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=20,test_size=0.2)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100, pr.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```

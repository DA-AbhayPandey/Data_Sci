```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(11,3))
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
#applying formula mean-3*standard_deviation to remove outliers 
min_range = dataset["CoapplicantIncome"].mean() - (3*dataset["CoapplicantIncome"].std())
max_range = dataset["CoapplicantIncome"].mean() + (3*dataset["CoapplicantIncome"].std())
min_range, max_range
```

```python
new_data=dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
plt.figure(figsize=(11,3))
sns.boxplot(x="CoapplicantIncome", data=new_data)
plt.show()
#we removed most of the outliers but still left with some
```

```python
new_data.shape
```

```python
z_score = (dataset["CoapplicantIncome"]-dataset["CoapplicantIncome"].mean())/dataset["CoapplicantIncome"].std()
z_score
```

```python
dataset["z_score"]=z_score
dataset[dataset["z_score"]<=3]
#we get the same result i.e 608 rows
```

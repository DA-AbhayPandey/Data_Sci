```python
import pandas as pd
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv("diabetes.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Outcome"] #Separate data in 2 parts x refers to input and y refers to outcomes
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
x.shape
```

```python
fs5 = SequentialFeatureSelector(lr,k_features=5,forward=True)
fs5.fit(x,y)
```

```python
 fs5.feature_names
```

```python
fs5.k_feature_names_
```

```python
fs5.k_score_
#Here we score of 5 features
```

```python
fs8 = SequentialFeatureSelector(lr,k_features=8,forward=True)
fs8.fit(x,y)
```

```python
 fs8.feature_names
```

```python
fs8.k_feature_names_
fs8.k_score_
#the score of feature 8
```

```python
#We do same for different feature No. to find out the beast selection

#for backward selection we use forward=False

fs3 = SequentialFeatureSelector(lr,k_features=3,forward=False)
fs3.fit(x,y)
```

```python
fs3.k_feature_names_
```

```python
fs3.k_score_
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
ar = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
pd.DataFrame(ar, columns=dataset.select_dtypes(include="float64").columns)
```

```python
new_dataset=pd.DataFrame(ar, columns=dataset.select_dtypes(include="float64").columns)
new_dataset.select_dtypes(include="float64").isnull().sum()
```
